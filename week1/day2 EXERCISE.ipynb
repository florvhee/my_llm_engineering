# Introducing the ollama package

And now we'll do the same thing, but using the elegant ollama python package instead of a direct HTTP call.

Under the hood, it's making the same call as above to the ollama server running at localhost:11434

```python
import ollama

response = ollama.chat(model=MODEL, messages=messages)
print(response['message']['content'])
```

## Alternative approach - using OpenAI python library to connect to Ollama

```python
# There's actually an alternative approach that some people might prefer
# You can use the OpenAI client python library to call Ollama:

from openai import OpenAI
ollama_via_openai = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')

response = ollama_via_openai.chat.completions.create(
    model=MODEL,
    messages=messages
)

print(response.choices[0].message.content)
```

## Are you confused about why that works?

It seems strange, right? We just used OpenAI code to call Ollama?? What's going on?!

Here's the scoop:

The python class `OpenAI` is simply code written by OpenAI engineers that makes calls over the internet to an endpoint.  

When you call `openai.chat.completions.create()`, this python code just makes a web request to the following url: "https://api.openai.com/v1/chat/completions"

Code like this is known as a "client library" - it's just wrapper code that runs on your machine to make web requests. The actual power of GPT is running on OpenAI's cloud behind this API, not on your computer!

OpenAI was so popular, that lots of other AI providers provided identical web endpoints, so you could use the same approach.

So Ollama has an endpoint running on your local box at http://localhost:11434/v1/chat/completions  
And in week 2 we'll discover that lots of other providers do this too, including Gemini and DeepSeek.

And then the team at OpenAI had a great idea: they can extend their client library so you can specify a different 'base url', and use their library to call any compatible API.

That's it!

So when you say: `ollama_via_openai = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')`  
Then this will make the same endpoint calls, but to Ollama instead of OpenAI.

# NOW the exercise for you

Take the code from day1 and incorporate it here, to build a website summarizer that uses Llama 3.2 running locally instead of OpenAI; use either of the above approaches.

```python
from bs4 import BeautifulSoup
import requests

def get_content_from_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.title.string if soup.title else 'No Title Found'
    for not_needing in soup(['script', 'style', "img", "input"]):
        not_needing.decompose()
    return title, soup.get_text()

class Website:
    def __init__(self, url):
        self.url = url
        self.title, self.content = get_content_from_url(url)

    def get_info(self):
        return f"Website Title: {self.title}, URL: {self.url}"
    
```

```python
website = Website("https://squareflow.be")
website.get_info()
```

```python
system_prompt = "You are a helpful assistant that extracts relevant information from a website to answer questions. Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Please answer in Markdown format."
def user_prompt_for(website):
    return f"Here is a website page {website.url} with title {website.title} and content: {website.content}\n\nBased on this information, please make a summary of the content of the page. It should provide the general information as well as the recent news if there are any."
```

```python
MODEL = "llama3.2"
def summarize_website(website):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
    response = ollama_via_openai.chat.completions.create(
        model=MODEL,
        messages=messages
    )
    completion = response.choices[0].message.content
    return completion
```

```python
summarize_website(website)
```

```python
def display_summary(url):
    scanned_website = Website(url)
    summary = summarize_website(scanned_website)
    display(Markdown(summary))

```

```python
display_summary("https://erebor.odoo.com")
```

```python
system_prompt = "You are an expert at marketing and branding and you are tasked with comparing websites. You will assess their strengths and weaknesses, you will rank them on a 10-scale and provide recommendations for improvements. Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Please answer in Markdown format."
def user_prompt_for(website1, website2):
    return f"Here are two websites for comparison:\n\nWebsite 1: {website1.url} - {website1.title}\nContent: {website1.content}\n\nWebsite 2: {website2.url} - {website2.title}\nContent: {website2.content}\n\nBased on this information, please compare the two websites and provide your analysis. Pick a winner and explain why."
```

```python
MODEL = "llama3.2"
def compare_websites(website1, website2):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website1, website2)}
    ]
    response = ollama_via_openai.chat.completions.create(
        model=MODEL,
        messages=messages
    )
    completion = response.choices[0].message.content
    return completion
```

```python
def compare_and_display(url1, url2):
    website1 = Website(url1)
    website2 = Website(url2)
    comparison = compare_websites(website1, website2)
    display(Markdown(comparison))
```

```python
compare_and_display("https://erebor.odoo.com", "https://squareflow.be")
```
